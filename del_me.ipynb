```python
%matplotlib inline
"""
    Footprint-detection pipeline schematic
    --------------------------------------
    • Works with matplotlib ≥ 3.4
    • Produces an 8-box left-to-right flow-chart

    Author: Deb the Data Science Guru
"""

import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# ---------------------------------------------------------------------
# 1.  Define the pipeline stages (label, x-centre position)
# ---------------------------------------------------------------------
STEPS = [
    ("Images\n(~90)",                     0),
    ("Group-strat\n5-fold split",         2),
    ("Resolution-aware\ntiling + cap",    4),
    ("High-conf\npseudo-labels",          6.5),
    ("Binary\nfootprint\ndetector",       9),
    ("Crop\nextraction",                 11),
    ("Species\nclassifier",              13),
    ("Metrics\nCV mAP / F1",             15),
]

# Box style
BOX_W   = 1.8          # width  (data units)
BOX_H   = 1.2          # height (data units)
Y_POS   = 0.5          # vertical centre
FONT_SZ = 10

# ---------------------------------------------------------------------
# 2.  Make the figure
# ---------------------------------------------------------------------
SAVE = False                           # set to True to write PNG
fig, ax = plt.subplots(figsize=(15, 3))
ax.axis('off')

# Draw boxes and text
for label, x in STEPS:
    rect = Rectangle((x - BOX_W/2, Y_POS - BOX_H/2),
                     BOX_W, BOX_H,
                     edgecolor='black', linewidth=1.5,
                     facecolor='none')
    ax.add_patch(rect)
    ax.text(x, Y_POS, label,
            ha='center', va='center',
            fontsize=FONT_SZ)

# Draw arrows between boxes
for idx in range(len(STEPS) - 1):
    x_start = STEPS[idx][1]   + BOX_W/2
    x_end   = STEPS[idx+1][1] - BOX_W/2
    ax.annotate(
        '', xy=(x_end,   Y_POS),
            xytext=(x_start, Y_POS),
            arrowprops=dict(arrowstyle='->', lw=1.5)
    )

fig.tight_layout()

if SAVE:
    fig.savefig("footprint_pipeline_schematic.png", dpi=300, bbox_inches='tight')
else:
    plt.show()
```
